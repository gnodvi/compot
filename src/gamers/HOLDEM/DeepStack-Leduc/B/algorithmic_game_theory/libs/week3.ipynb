```python
%load_ext autoreload
%autoreload 2

import numpy as np
import libs.week1 as week1
import libs.week2 as week2
import libs.week3 as week3
import matplotlib.pyplot as plt
```

```python
rps_matrix = np.array([[-1, 0, 1], [1, -1, 0], [0, 1, -1]])
non_rps_matrix = np.array([[1, 2, 3], [-1, -2, -1], [3, 1, -4]])
```

```python
matrix = non_rps_matrix
fict_play = week3.FictitiousPlay(matrix=matrix)

iterations = 100
exploitabilities = []

for i in range(iterations):
    fict_play.iteration()

    exploitability = week2.compute_exploitability_zero_sum(matrix=matrix,
                                                           row_strategy=fict_play.average_row_strategy,
                                                           column_strategy=fict_play.average_column_strategy)
    exploitabilities.append(exploitability)

plt.plot(range(iterations), exploitabilities)
```
